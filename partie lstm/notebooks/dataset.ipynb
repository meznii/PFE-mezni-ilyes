```python
def transform_data(data, number_of_samples, number_of_result):
    df= pd.DataFrame({'history': [],  'predict': []  })
    features = data['cons_glbal'].tolist()
    list_devices = list(data["id_device"])
    for device in devices :
        start_index = list_devices.index(device)
        for i in range (0,devices[device]-number_of_result,number_of_samples):
            history = features[start_index+i:start_index+i+number_of_samples]
            predict = features[start_index+i+number_of_samples:start_index+i+number_of_samples+number_of_result]
            df = df.append({'history': history , 'predict': predict} , ignore_index=True)
    return df
```

```python
def transform_data(data, number_of_samples, number_of_result):
    features = data['cons_glbal'].values.reshape(-1,1)
    X, Y = [], []
    list_devices = list(data["id_device"])
    for device in devices :
        start_index = list_devices.index(device)
        for i in range (0,devices[device]-number_of_samples,number_of_samples):
            X.append(features[start_index+i:start_index+i+number_of_samples,0])
            Y.append(features[start_index+i+number_of_samples:start_index+i+number_of_samples+number_of_result,0])

    return X, Y
```

```python
X, Y = transform_data(dataset, 50,1)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
model = Sequential()
model.add(LSTM(50, input_shape=(50, 1), activation="tanh"))
model.add(Dense(1))
model.compile(loss='MSE', optimizer='adam')
model.fit(X, Y, epochs=100, batch_size=1, verbose=True)
```

```python
def transform_data_mois(data, number_of_samples, number_of_result):
    features = data['cons_glbal'].values.reshape(-1,1)
    month_liste = dict(data["month"].value_counts())
    X, Y = [], []
    list_month = list(data["month"])
    for month in month_liste :
        start_index = list_month.index(month)
        for i in range (0,month_liste[month]-number_of_samples,number_of_samples):
            X.append(features[start_index+i:start_index+i+number_of_samples,0])
            Y.append(features[start_index+i+number_of_samples:start_index+i+number_of_samples+number_of_result,0])

    return X, Y
```
